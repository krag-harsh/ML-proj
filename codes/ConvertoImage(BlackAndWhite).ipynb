```python
from torchaudio import transforms
import torchaudio
import numpy as np
import librosa
import skimage.io
```

```python
def openAudio(audioPath):
  y,sr = librosa.load(audioPath,offset = 1.0,duration = 10.0,sr = 22050)
  return (y,sr)
```

```python
def scale(X,min,max):
  X_std = (X-X.min()) / (X.max() - X.min())
  X_scaled = X_std * (max - min) + min
  return X_scaled
```

```python
def spectrogram_Save(y,sr,out,hop_length,n_mels):
  spec = librosa.feature.melspectrogram(y=y,sr=sr,n_mels = n_mels,n_fft = hop_length*2,hop_length = hop_length)

  spec = np.log(spec + 1e-9)

  image = scale(spec,0,255).astype(np.uint8)
  image = np.flip(image,axis = 0)
  image = 255 - image
  skimage.io.imsave(out,image)
```

```python
import os
folderpath="/content/drive/MyDrive/ML-project/dataTop5/"
cl=os.listdir(folderpath)
for c in cl:
  fpath=folderpath+c
  li=os.listdir(fpath)
  for aud in li:
    audioPath = fpath+"/"+aud
    try:
      audio = openAudio(audioPath)
      y,sr = audio
      out = "/content/drive/MyDrive/ML-project/dataTop5_image/"+c+"/"+aud[:-3]+"png"
      hop_length = 512
      n_mels = 128
      time_steps = 384
      start = 0
      length = time_steps*hop_length 
      window = y[start:start+length]
      spectrogram_Save(window,sr=sr,out = out,hop_length = hop_length,n_mels = n_mels)
    except:
      print("audiopath=",audioPath)
```
